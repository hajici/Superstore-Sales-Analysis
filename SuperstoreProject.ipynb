```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.ticker as mticker
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('/Users/hajimeci/Desktop/Data Analysis/Superstore.csv')
df.head()
```

```python
# Dropping Row ID column.
df.drop(columns=['Row ID'])
```

```python
df.info()
```

```python
# Changing column names.
df.columns = ['row_id','order_id','order_date','ship_date','ship_mode','customer_id','customer_name',
                    'segment','country','city','state','postal_code','region','product_id',
                    'category','sub_category','product_name','sales']
```

```python
df.head()
```

```python
df.info()
```

```python
# Changing Row ID column to string type.
df['row_id'] = df['row_id'].astype(str)
```

```python
# Changing Order Date and Ship Date to pandas datetime format.
df['order_date'] = pd.to_datetime(df['order_date'], format='%d/%m/%Y')
df['ship_date'] = pd.to_datetime(df['ship_date'], format='%d/%m/%Y')
```

```python
# I notice there are null values under the Postal Code column, let's take a look to these values.
df[df['postal_code'].isnull()]
```

We are only missing the Postal Codes of Burlington in the State of Vermont.

```python
# Let's fill these colums with their Postal Code.
df['postal_code'] = df['postal_code'].fillna(5401)
```

### Exploratory Data Analysis (EDA)

```python
sns.set_palette('pastel')
```

```python
# Let's take a look to the distribution of some columns.
fig, [[ax1, ax2, ax3], [ax4, ax5, ax6]] = plt.subplots(2,3, figsize=(15,12))
sns.countplot(ax=ax1, data=df, x='ship_mode')
sns.histplot(ax=ax2, data=df, x='ship_date', kde=True)
sns.countplot(ax=ax3, data=df, x='segment')
sns.countplot(ax=ax4, data=df, x='region')
sns.countplot(ax=ax5, data=df, x='category')
sns.countplot(ax=ax6, data=df, x='sub_category')


ax1.tick_params(axis='x', rotation=90)
ax2.tick_params(axis='x', rotation=90)
ax6.tick_params(axis='x', rotation=90)
fig.subplots_adjust(top=10)
fig.tight_layout()
plt.show()
```

## Observations:
1. The prefered Ship Mode is Standard Class.
2. Ship dates are between January 2015 and January 2019.
3. Most sales come from the Consumer Segment.
4. Sales by Region are lead by the West, followed by the East, Central and South.
5. Office Supplies is the leading Category in sales.
6. The best 2 selling Sub-Category is Binders and Paper.

```python
# Creating columns for analysis.
df['order_day'] = df.order_date.dt.day_name()
df['ship_day'] = df.ship_date.dt.day_name()
df['delivery_days'] = (df.ship_date - df.order_date).dt.days
```

```python
fig, ax = plt.subplots()
sns.histplot(data=df, x='order_date', bins=50, kde=True)
ax.tick_params(axis='x', rotation=90)
plt.show()
```

There is an intereting pattern in sales. I am going to look deeper into it.

```python
# Creating subsets to analyze data by year.
od_15 = df.loc[df['order_date'].dt.year == 2015]
od_16 = df.loc[df['order_date'].dt.year == 2016]
od_17 = df.loc[df['order_date'].dt.year == 2017]
od_18 = df.loc[df['order_date'].dt.year == 2018]
```

```python
fig, [[ax1, ax2], [ax3, ax4]] = plt.subplots(2,2)
sns.histplot(ax=ax1, data=od_15, x='order_date', bins=12)
sns.histplot(ax=ax2, data=od_16, x='order_date', bins=12)
sns.histplot(ax=ax3, data=od_17, x='order_date', bins=12)
sns.histplot(ax=ax4, data=od_18, x='order_date', bins=12)


ax1.tick_params(axis='x', rotation=90)
ax2.tick_params(axis='x', rotation=90)
ax3.tick_params(axis='x', rotation=90)
ax4.tick_params(axis='x', rotation=90)
fig.subplots_adjust(top=10)
fig.tight_layout()
plt.show()
```

### Observations:
1. The mayority of orders are made in September, November and December every year.
2. January and February tend to be low months for orders.

```python
fig, ax = plt.subplots()
sns.countplot(data=df, x='order_day', order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

ax.tick_params(axis='x', rotation=90)
plt.show()
```

Tuesday, saturday, sunday and monday seems to be the prefer days to order. In other hand, Thursday shows a significantly lower amount of orders compared to other days of the week.


```python
fig, ax = plt.subplots()
sns.countplot(data=df, x='ship_day', order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

ax.tick_params(axis='x', rotation=90)
plt.show()
```

Ship days are evenly distributed with highest on thursday and a lowest on monday.

### Correlation of Features

```python
# Correlation by Ship Mode.
ship = df[['ship_mode', 'sales']]
ship_cat = pd.get_dummies(ship)
ship_mat = ship_cat.corr()

fig.set_size_inches(10,5)
sns.heatmap(data = ship_mat, square = True, annot = True, cbar = True)
plt.show()
```

```python
# Correlation by Segment.
seg = df[['segment', 'sales']]
seg_cat = pd.get_dummies(seg)
seg_mat = seg_cat.corr()

fig.set_size_inches(10,5)
sns.heatmap(data = seg_mat, square = True, annot = True, cbar = True)
plt.show()
```

```python
# Correlation by Category.
cat = df[['category', 'sales']]
cat_cat = pd.get_dummies(cat)
cat_mat = cat_cat.corr()

fig.set_size_inches(10,5)
sns.heatmap(data = cat_mat, square = True, annot = True, cbar = True)
plt.show()
```
